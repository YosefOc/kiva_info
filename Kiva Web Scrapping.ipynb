```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.error import HTTPError
try: 
    url = "https://www.kiva.org/lend?status=funded" # Scrape the HTML using the url
    r = requests.get(url)        #this is where the HTML Markup will be stored in response object
    soup = BeautifulSoup(r.text, "lxml")
    print(soup.h6)
except HTTPError as he: 
    print(he)
    #exit()
    #continue 
except AttibuteError as ae: 
    print(ae)
    exit()
```

```python
print(soup.prettify())
```

```python
# find all the tags 
for tag in soup.find_all(True):
    print(tag.name)
```
